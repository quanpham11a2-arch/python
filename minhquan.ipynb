## [1단계] 난장판(더미) 파일 준비하기
먼저, 청소할 가짜 쓰레기 파일들이 필요합니다. 코랩 환경 내에 임의의 파일 100개를 무작위로 생성하는 코드를 제미니에게 물어보고, 아래 셀에 붙여넣어 실행하세요.

* 💡 **프롬프트 팁**: "코랩 환경의 'downloads'라는 폴더 안에 임의의 가짜 파일 100개를 만들어주는 파이썬 코드를 짜줘. 확장자는 pdf, docx, jpg, png, txt, xlsx가 섞여 있게 해줘."

```python
# 여기에 제미니가 작성해 준 '더미 파일 생성 코드'를 붙여넣고 실행하세요.
```

```python
# Dán và chạy 'mã tạo tệp giả' do Gemini tạo vào đây.
import os
import random
import string

# Tạo thư mục downloads
folder_name = "downloads"
os.makedirs(folder_name, exist_ok=True)

# Danh sách các đuôi tệp
extensions = ["pdf", "docx", "jpg", "png", "txt", "xlsx"]

# Tạo 100 tệp giả
for i in range(100):

    # Tạo tên tệp ngẫu nhiên
    random_name = ''.join(random.choices(string.ascii_lowercase, k=8))

    # Chọn ngẫu nhiên một đuôi tệp
    ext = random.choice(extensions)

    # Hoàn thành tên tệp
    file_name = f"{random_name}.{ext}"

    # Đường dẫn tệp
    file_path = os.path.join(folder_name, file_name)

    # Tạo tệp rỗng
    with open(file_path, "w") as f:
        f.write("dummy file")

print("Đã tạo 100 tệp giả!")
```

## [2단계] 확장자별 자동 분류 스크립트 작성
이제 위에서 만든 수많은 파일들을 확장자별로 새 폴더를 만들어 깔끔하게 이동시켜주는 분류 코드를 제미니에게 요청하고 아래 셀에 붙여넣어 실행하세요.

* 💡 **프롬프트 팁**: "'downloads' 폴더 안에 있는 파일들을 확장자별로 확인해서, 확장자 이름과 똑같은 새 폴더를 만들고 그 안으로 파일들을 모두 이동시켜주는 파이썬 코드를 만들어 줘. 이미 폴더가 있으면 만들지 말고 그냥 이동시켜 줘."

```python
# 여기에 제미니가 작성해 준 '파일 자동 분류 코드'를 붙여넣고 실행하세요.
```

```python
import os
import shutil

# đường dẫn thư mục downloads
folder_path = "downloads"

# kiểm tra các file trong thư mục downloads
for filename in os.listdir(folder_path):

    # đường dẫn đầy đủ của file
    file_path = os.path.join(folder_path, filename)

    # kiểm tra xem có phải là file không
    if os.path.isfile(file_path):

        # lấy phần đuôi file (extension)
        extension = filename.split(".")[-1]

        # tạo đường dẫn thư mục mới theo tên extension
        target_folder = os.path.join(folder_path, extension)

        # nếu thư mục chưa tồn tại thì tạo mới
        os.makedirs(target_folder, exist_ok=True)

        # di chuyển file vào thư mục tương ứng
        shutil.move(file_path,
                    os.path.join(target_folder, filename))

print("Đã hoàn thành tự động phân loại file!")
```

```python
import shutil
import os

folder_to_delete = "downloads"

# Kiểm tra xem thư mục có tồn tại không trước khi xóa
if os.path.exists(folder_to_delete):
    shutil.rmtree(folder_to_delete)
    print(f"Thư mục '{folder_to_delete}' đã được xóa thành công.")
else:
    print(f"Thư mục '{folder_to_delete}' không tồn tại.")
```
