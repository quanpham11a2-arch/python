```python
import csv
import random
import time
import os

# ======================
# CONFIGURATION
# ======================
filename = "my_words.csv"
score_file = "highscore.txt"
words = []
missed_words = []  # Danh sách lưu từ sai để ôn tập TOPIK

# ======================
# STEP 1: LOAD DATA
# ======================
try:
    with open(filename, 'r', encoding='utf-8-sig') as f:
        reader = csv.reader(f)
        next(reader, None)
        for row in reader:
            if len(row) >= 2:
                words.append({
                    '한국어': row[0].strip(),
                    '뜻': row[1].strip()
                })
except FileNotFoundError:
    print(f"❌ Lỗi: Không tìm thấy file '{filename}'. Hãy upload file vào thư mục bên trái!")
    words = []

# ======================
# STEP 2: LOAD HIGH SCORE
# ======================
if os.path.exists(score_file):
    with open(score_file, "r") as f:
        try:
            high_score = int(f.read())
        except:
            high_score = 0
else:
    high_score = 0

# ======================
# STEP 3: GAME LOGIC (MAX 10 QUESTIONS)
# ======================
if len(words) >= 4: # Cần ít nhất 4 từ để tạo trắc nghiệm A/B/C/D
    score = 0
    correct_count = 0

    # Shuffle và lấy đúng 10 câu (hoặc ít hơn nếu file không đủ 10 từ)
    quiz_list = words.copy()
    random.shuffle(quiz_list)
    num_questions = min(10, len(quiz_list))

    print("="*40)
    print(f"🎮 CHƯƠNG TRÌNH LUYỆN THI TOPIK: 10 CÂU NGẪU NHIÊN")
    print(f"🥇 Điểm cao nhất hiện tại: {high_score}")
    print("="*40)

    for i, q in enumerate(quiz_list[:num_questions]):
        correct_answer = q['뜻']

        # Tạo 3 đáp án sai ngẫu nhiên
        wrong_candidates = [w['뜻'] for w in words if w['뜻'] != correct_answer]
        wrong_answers = random.sample(wrong_candidates, 3)

        options = wrong_answers + [correct_answer]
        random.shuffle(options)

        print(f"\n📝 Câu {i+1}/{num_questions}: Từ '{q['한국어']}' nghĩa là gì?")
        print("⏳ Bạn có 15 giây để phản xạ!")

        labels = ['A', 'B', 'C', 'D']
        for j in range(4):
            print(f"{labels[j]}. {options[j]}")

        start_time = time.time()
        user_input = input("👉 Chọn A/B/C/D (hoặc '종료' để thoát): ").strip().upper()
        end_time = time.time()

        if user_input == '종료':
            print("\n👋 Đã thoát chương trình ôn tập.")
            break

        elapsed = end_time - start_time

        # Kiểm tra đáp án
        if user_input in labels:
            selected_answer = options[labels.index(user_input)]

            if selected_answer == correct_answer:
                if elapsed <= 15:
                    print(f"✅ CHÍNH XÁC! ({elapsed:.1f}s)")
                    score += 1
                    correct_count += 1
                else:
                    print(f"⏰ ĐÚNG, nhưng quá thời gian ({elapsed:.1f}s). Cần phản xạ nhanh hơn!")
                    missed_words.append({'từ': q['한국어'], 'nghĩa': correct_answer, 'lỗi': 'Quá giờ'})
            else:
                print(f"❌ SAI RỒI! Đáp án đúng phải là: {correct_answer}")
                missed_words.append({'từ': q['한국어'], 'nghĩa': correct_answer, 'lỗi': 'Sai nghĩa'})
        else:
            print("⚠️ Lệnh không hợp lệ. Coi như bỏ lỡ câu này.")
            missed_words.append({'từ': q['한국어'], 'nghĩa': correct_answer, 'lỗi': 'Nhập sai định dạng'})

    # ======================
    # STEP 4: FINAL RESULT & RANKING
    # ======================
    print("\n" + "="*40)
    print("📊 TỔNG KẾT BUỔI HỌC")
    print(f"✅ Số câu đúng trong thời gian: {correct_count}/{num_questions}")
    print(f"🏆 Tổng điểm tích lũy: {score}")

    # Xếp hạng dựa trên điểm TOPIK
    if score == 10:
        rank = "💎 CẤP ĐỘ 6 (Xuất sắc!)"
    elif score >= 7:
        rank = "🥇 CẤP ĐỘ 4-5 (Rất tốt)"
    elif score >= 4:
        rank = "🥈 CẤP ĐỘ 2-3 (Tạm ổn)"
    else:
        rank = "🥉 CẤP ĐỘ 1 (Cần cố gắng nhiều hơn)"
    print(f"🎖️ Xếp hạng năng lực: {rank}")

    # Hiển thị từ cần học lại
    if missed_words:
        print("\n📚 DANH SÁCH TỪ CẦN ÔN LẠI:")
        for m in missed_words:
            print(f"📍 {m['từ']}: {m['nghĩa']} ({m['lỗi']})")

    # Lưu High Score
    if score > high_score:
        with open(score_file, "w") as f:
            f.write(str(score))
        print("\n🔥 KỶ LỤC MỚI ĐÃ ĐƯỢC THIẾT LẬP!")

else:
    print("❌ Dữ liệu không đủ để tạo bài trắc nghiệm. Cần ít nhất 4 từ trong file CSV.")
```
