```python
import csv
import random
import time

filename = "my_words.csv"
words = []

try:
    with open(filename, 'r', encoding='utf-8-sig') as f:
        reader = csv.reader(f)

        for row in reader:
            if len(row) >= 2:
                words.append({
                    '한국어': row[0].strip(),
                    '뜻': row[1].strip()
                })

    print(f"🎉 Thành công! Đã tải {len(words)} từ vựng.")

except FileNotFoundError:
    print(f"⚠️ Không tìm thấy file '{filename}'.")
    words = []

# =========================
# GAME QUIZ
# =========================
if len(words) > 0:

    while True:

        question = random.choice(words)
        korean_word = question['한국어']
        meaning = question['뜻']

        print(f"\n📝 Câu hỏi: '{korean_word}' nghĩa là gì? (⏳ giới hạn 15 giây)")

        # bắt đầu tính giờ
        start_time = time.time()

        # nhập câu trả lời
        user_answer = input("👉 Nhập đáp án (gõ '종료' để thoát): ")

        # nếu thoát
        if user_answer == '종료':
            print("👋 Kết thúc quiz. Cảm ơn bạn!")
            break

        # kết thúc tính giờ
        end_time = time.time()
        elapsed_time = end_time - start_time

        # xử lý dữ liệu
        user_answer = user_answer.strip()
        correct_answer = meaning.strip()

        # =========================
        # KIỂM TRA ĐÁP ÁN + THỜI GIAN
        # =========================
        if user_answer == correct_answer and elapsed_time <= 15:
            print(f"✅ Đúng rồi! (⏱ {elapsed_time:.1f} giây)")
        elif user_answer == correct_answer and elapsed_time > 15:
            print(f"⏰ Đúng nhưng quá thời gian! ({elapsed_time:.1f} giây)")
        else:
            print(f"❌ Sai rồi! Đáp án đúng là: {correct_answer}")

else:
    print("❌ Không có dữ liệu từ vựng.")
```
